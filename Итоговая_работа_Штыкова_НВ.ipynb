```python
#import csv
import pandas as pd
import numpy as np
#1 скачать набор данных

diamonds = pd.read_csv('/content/diamonds.csv')
display(diamonds)

#2 Узнать  основные сведения о данных, изучите типы данных, названий колонок, количество строк и столбцов
diamonds.info()

#3  Проверьте данные на наличие пропусков в полях
d_N=diamonds.isna().sum()
print(d_N)


#4 Заполните пропуски 0 в полях с числовыми значениями
diamonds.fillna('0')
display(diamonds.fillna('0'))

#5 Сгруппируйте данные по качеству алмазов и посчитайте:
#a  количество алмазов
display(diamonds.groupby('cut')['carat'].count())

#b среднюю цену
avg_price = np.round((diamonds.groupby('cut')['price'].mean()), 2)
display(avg_price)

#c  максимальный и минимальный карат
print(diamonds.groupby('cut')['carat'].max())
print(diamonds.groupby('cut')['carat'].min())

#6  На основе этого сформулируйте выводы:
#a  Алмазов какого качества больше всего в данных
print(diamonds.groupby('cut')['carat'].count())

#b Какая разница между средними ценами на алмазы качества ideal  и premium
price_id = np.round(diamonds.query('cut=="Ideal"')['price'].mean(), 2)
print(price_id)
price_prem = np.round(diamonds.query('cut=="Premium"')['price'].mean(), 2)
print(price_prem)
sub_price= np.round((price_prem - price_id), 2)
print(sub_price)

#7 Постройте сводную таблицу по цвету и качеству посчитав средний вес алмазов
display(diamonds.pivot_table(
        index='color',
        columns = 'cut',
        values = 'carat',
        aggfunc = 'mean'
    ))

#8 Отфильтруйте алмазы с качеством good. Найдите их среднее значение высоты, ширины и глубины
g_d=diamonds.query('cut == "Good"')
display(g_d)
display(g_d.groupby('cut')['x'].mean())
display(g_d.groupby('cut')['y'].mean())
display(g_d.groupby('cut')['z'].mean())
```
